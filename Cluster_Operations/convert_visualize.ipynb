```python
import csv
```

```python
clstr_file = '/file.clstr'
```

```python
clusters = {}
current_cluster_id = None
```

```python
with open(clstr_file, 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith('>Cluster'):
            current_cluster_id = int(line.split(' ')[1])
            clusters[current_cluster_id] = []
        else:
            sequence_id = line.split('>')[1].split('...')[0]
            clusters[current_cluster_id].append(sequence_id)
```

```python
csv_file = '/cluster_file.csv'
```

```python
with open(csv_file, 'w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['Cluster ID', 'Sequence ID'])

    for cluster_id, sequence_ids in clusters.items():
        for sequence_id in sequence_ids:
            csv_writer.writerow([cluster_id, sequence_id])
```

```python
print("Conversion to CSV completed successfully.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_file = '/cluster_file.csv'
```

```python
cluster_sizes = []
```

```python
with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        cluster_sizes.append(int(row[0]))
```

```python
sns.histplot(cluster_sizes, bins=20, kde=True)
plt.xlabel('Cluster Size')
plt.ylabel('Frequency')
plt.title('Distribution of Cluster Sizes')
plt.show()
```

```python
with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        cluster_id = int(row[0])
        seq_id = row[1]
        if cluster_id not in clusters:
            clusters[cluster_id] = []
        clusters[cluster_id].append(seq_id)

```

```python
for cluster_id, seq_ids in clusters.items():
    print(f"Cluster {cluster_id}: {seq_ids}")

```
