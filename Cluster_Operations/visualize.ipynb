```python
import matplotlib.pyplot as plt
```

```python
data = []
```

```python
with open('/content/clstr_size.txt', 'r') as file:
    lines = file.readlines()
    is_cluster_distribution = False
    for line in lines:
        line = line.strip()
        if line == 'Cluster size distribution:':
            is_cluster_distribution = True
            continue
        if is_cluster_distribution and ':' in line:
            cluster_size, sequence_count = line.split(':')
            cluster_size = cluster_size.strip()
            sequence_count = sequence_count.strip()
            if cluster_size.isnumeric() and sequence_count.isnumeric():
                data.append([int(cluster_size), int(sequence_count)])
```

```python
x = [entry[0] for entry in data]
y = [entry[1] for entry in data]
```

```python
plt.scatter(x, y)
plt.xlabel('Cluster Size')
plt.ylabel('Sequence Count')
plt.title('Cluster Size Distribution')
plt.show()
```
